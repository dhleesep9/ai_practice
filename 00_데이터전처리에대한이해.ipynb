```python

fish_length = [25.4, 26.3, 26.5, 29.0, 29.0, 29.7, 29.7, 30.0, 30.0, 30.7, 31.0, 31.0,
                31.5, 32.0, 32.0, 32.0, 33.0, 33.0, 33.5, 33.5, 34.0, 34.0, 34.5, 35.0,
                35.0, 35.0, 35.0, 36.0, 36.0, 37.0, 38.5, 38.5, 39.5, 41.0, 41.0, 9.8,
                10.5, 10.6, 11.0, 11.2, 11.3, 11.8, 11.8, 12.0, 12.2, 12.4, 13.0, 14.3, 15.0]
fish_weight = [242.0, 290.0, 340.0, 363.0, 430.0, 450.0, 500.0, 390.0, 450.0, 500.0, 475.0, 500.0,
                500.0, 340.0, 600.0, 600.0, 700.0, 700.0, 610.0, 650.0, 575.0, 685.0, 620.0, 680.0,
                700.0, 725.0, 720.0, 714.0, 850.0, 1000.0, 920.0, 955.0, 925.0, 975.0, 950.0, 6.7,
                7.5, 7.0, 9.7, 9.8, 8.7, 10.0, 9.9, 9.8, 12.2, 13.4, 12.2, 19.7, 19.9]

import numpy as np
data = np.column_stack((fish_length, fish_weight))
answer = np.concatenate((np.ones(35), np.zeros(14)))
from sklearn.model_selection import train_test_split
train_data, test_data, train_answer, test_answer = train_test_split(data, answer, stratify=answer, random_state=42)
print(test_answer)

from sklearn.neighbors import KNeighborsClassifier
kn = KNeighborsClassifier()
kn.fit(train_data, train_answer)
kn.score(test_data, test_answer)
import matplotlib.pyplot as plt
distances, indexes =kn.kneighbors([[25,100]])
plt.scatter(train_data[:,0],train_data[:,1])
plt.scatter(25, 150, marker='^')
plt.scatter(train_data[indexes,0], train_data[indexes,1], marker='^')
plt.xlim((0,1000))
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```

```python
mean = np.mean(data, axis=0)
std = np.std(data, axis=0)
data_scaled= (data-mean)/std
train_data, test_data, train_answer, test_answer = train_test_split(data_scaled, answer, stratify=answer, random_state=42)
kn.fit(train_data, train_answer)
kn.score(test_data, test_answer)
plt.scatter(train_data[:,0],train_data[:,1])
tmp=([25, 100]-mean)/std
distances, indexes =kn.kneighbors([tmp])
plt.scatter(tmp[0],tmp[1], marker='^')
plt.scatter(train_data[indexes,0], train_data[indexes,1], marker='^')
plt.xlabel('length')
plt.ylabel('weight')
plt.show()
```
